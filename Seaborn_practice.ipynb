```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ir = sns.load_dataset("iris")
```

```python
ir.info()
```

```python
#scatterplot
sns.scatterplot( data=ir)
plt.title("Scatter Plot")
plt.show()

#lineplot
sns.lineplot(data = ir)
plt.title("Line Plot")
plt.show()

#barplot
sns.barplot(data = ir)
plt.title("Bar Plot")
plt.show()

#countplot
sns.countplot(data = ir)
plt.title("Count Plot")
plt.show()


```
