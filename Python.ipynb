```python
#program on palindrome
num = 123
temp = num
reverse = 0
while num>0:
  reminder =num%10
  reverse = reverse*10+reminder
  num = num//10
print(reverse)
if(temp==reverse):
  print("palindrome")
else:
  print("Not a palindrome")
```

```python
#Armstrong number
num = 944
sum = 0
temp = num
l = len(str(num))
#print(l)
for i in range(l+1):
  rem = num%10
  sum = sum+(rem**l)
  num = num//10
print(sum)
if(temp==sum):
  print("Armstrong")
else:
  print("Not a armstrong")
```

3. Set

Definition

A Set is an unordered collection of unique elements.

Features Unordered Mutable No duplicates allowed Faster searching

Important Functions Function Description

>add() Add element

>remove() Remove element

>union() Combine sets

>intersection() Common elements

>difference() Unique elements

>clear() Remove all elements


```python
#Sets
Alpha = ("A","B","C","D")
print(Alpha)

name1={"a","b","c","d"}
name2={"d","e","f","b"}
#Add
name1.add("g")
print(name1)
#Remove
name1.remove("a")
print(name1)
#union
print(name1.union(name2))
#intersection
print(name1.intersection(name2))
#difference
print(name1.difference(name2))
#clear
name1.clear()
print(name1)
```

```python
#list
students = []

marks = {}
name = input("Enter student name: ")
age = int(input("Enter age: "))
n = int(input("Enter number of subjects: "))

students.append(name)
student_info = (name, age)
for i in range(n):
    sub = input("Enter subject name: ")
    mark = int(input("Enter marks: "))

    subjects.add(sub)
    marks[sub] = mark

total = sum(marks.values())
average = total / n
print("\n----- REPORT CARD -----")
print("Name:", name)
print("Age:", age)
print("Student List:", students)
print("Student Info:", student_info)
print("Subjects:", subjects)
print("Marks:", marks)
print("Total:", total)
print("Average:", average)
```
