```python
import pandas as pd

data = pd.read_csv('netflix_titles.csv')

data.head()
```

# New Section

```python
print(data['type'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='type')
plt.title('Netflix Content Type')
plt.show()
```

```python
top_countries = data['country'].value_counts().head(10)

print(top_countries)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries Contributing Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
top_directors = data['director'].dropna().value_counts().head(10)
print(top_directors)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values,y=top_directors.index)
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
#Functions
#scatterplot
plt.figure(figsize=(6,4))
temp = data.head(10)
sns.scatterplot(data=temp,x='release_year',y='date_added')
plt.title("Scatter plot - Release Year vs Size")
plt.show()



#Line plot
plt.figure(figsize=(6,4))
sns.lineplot(x="director", y="type", data=temp)
plt.title("Line Plot - Total Bill vs Size")
plt.show()

```
