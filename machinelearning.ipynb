```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#simple linear
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Marks":[20,35,50,65,80,95]
})
print(df)
X = df[["Hours"]]
y = df["Marks"]
model = LinearRegression()
model.fit(X,y)
prediction = model.predict([[3.5]])
print("Predicted Marks:", prediction[0])
```

```python
#Multiple Linear Regression

df = pd.DataFrame({
    "Hours":[1,2,3,4,5],
    "Assignments":[2,4,6,8,10],
    "Marks":[35,50,65,80,95]
})

plt.scatter(df["Hours"], df["Marks"])

plt.title("Multiple Linear Regression")
plt.xlabel("Hours")
plt.ylabel("Marks")

plt.show()
```

```python
#Logistic Regression
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Hours":[1,2,3,4,5,6],
    "Pass":[0,0,0,1,1,1]
})

plt.scatter(df["Hours"], df["Pass"])

plt.title("Logistic Regression")
plt.xlabel("Hours")
plt.ylabel("Pass")

plt.show()
```

```python
#Decision Tree
x=[1,2,3,4,5]
y=[0,0,1,1,1]

plt.step(x,y)

plt.title("Decision Tree")

plt.xlabel("Feature")

plt.ylabel("Class")
```

```python
#Random Forest
trees=["T1","T2","T3","T4","Final"]
accuracy=[80,82,85,84,88]

plt.bar(trees,accuracy)

plt.title("Random Forest")

plt.xlabel("Trees")

plt.ylabel("Accuracy")

plt.show()
```

```python
#K-Nearest Neighbors (KNN)
x=[1,2,3,4,5]
y=[1,2,2,3,3]
plt.scatter(x,y)
plt.title("KNN")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[0,0,1,1,1]
plt.scatter(x,y)
plt.axvline(x=3,color="red")
plt.title("Support Vector Machine")
plt.xlabel("Feature")
plt.ylabel("Class")
plt.show()
```

```python
#Naive Bayes
classes=["Yes","No"]
probability=[0.7,0.3]
plt.bar(classes,probability)
plt.title("Naive Bayes")
plt.xlabel("Class")
plt.ylabel("Probability")
plt.show()
```
