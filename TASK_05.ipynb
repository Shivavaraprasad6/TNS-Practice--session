```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Book_ID':[101,102,103,104,105],
    'Book_Name':['Python','DS','AI','Machine Learning', 'Database Systems'],
    'Category':['Programming','Programming','AI','AI','DBMS'],
    'Borrow_Count':[150,200,180,220,120],
    'Rating':[4.5,4.8,4.7,4.9,4.3]
}

df = pd.DataFrame(data)

df
```

```python

df.describe()
```

```python
#Handle Missing Values
df.isnull().sum()
```

```python
df['Borrow_Count'] = df['Borrow_Count'].fillna(df['Borrow_Count'].mean())
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
df
```

```python
avg_borrow = df['Borrow_Count'].mean()

print(avg_borrow)
```

```python
#Find Books Borrowed More Than Average
avg_borrow = df['Borrow_Count'].mean()
df[df['Borrow_Count'] > avg_borrow]
```

```python
#Sort by Borrow Count
df.sort_values(by='Borrow_Count', ascending=False)
```

```python
#Find Category-wise Average Ratings
df.groupby('Category')['Rating'].mean()
```

```python
#Create Popularity Index
df['Popularity_Index'] = (df['Borrow_Count'] * df['Rating']) / 100

df
```

```python
#numpy functions
np.max(df['Borrow_Count'])
```

```python
np.min(df['Borrow_Count'])
```

```python
np.mean(df['Rating'])
```

```python
np.std(df['Rating'])
```

```python
#Identify the Most Borrowed Book
df.loc[df['Borrow_Count'].idxmax()]
```

```python
#Generate Report
print("===== LIBRARY REPORT =====")

print("Total Books:", len(df))

print("Average Borrow Count:",
      df['Borrow_Count'].mean())

print("Most Borrowed Book:",
      df.loc[df['Borrow_Count'].idxmax(),'Book_Name'])

print("Highest Borrow Count:",
      df['Borrow_Count'].max())

print("\nCategory-wise Average Ratings")
print(df.groupby('Category')['Rating'].mean())
```
